# Pandas Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r'C:\Users\Nosa-Ihaza Oduwa\Documents\Ice Cream Rating.xlsx')
df = df.set_index('Date')
df
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores' )
```

```python
df.plot.barh( stacked = True)
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 600, c = 'Yellow')
```

```python
df.plot.hist(bins = 20)
```

```python
df.plot.box()
```

```python
df.boxplot()
```

```python
df.plot.area(figsize = (10,6))
```

```python
df.plot.pie(y ="Texture Rating", figsize = (10,10))
```
