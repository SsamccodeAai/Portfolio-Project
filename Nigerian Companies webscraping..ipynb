# Webscraping List Of Largest Companies Nigerian Using Python(Pandas library)

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_companies_of_Nigeria'
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
print(soup)
```

```python
soup.find('table')
```

```python
table = soup.find_all('table')[1]
table
```

```python
Comp_titles_list = table.find_all('th')
print(Comp_titles_list)
```

```python
Comp_table_list = [title.text.strip() for title in Comp_titles_list]
print(Comp_table_list)
```

```python
#importing pandas library
```

```python
import pandas as pd
df = pd.DataFrame(columns = Comp_table_list)
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_table_data = [data.text.strip() for data in row_data ]
   
    lenght = len(df)
    df.loc[lenght] = individual_table_data
    
```
